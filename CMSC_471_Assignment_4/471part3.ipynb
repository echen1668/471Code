```python
# Random Forest using 'titanic_train.csv'
import numpy as np
import pandas as pd
#Load data
df = pd.read_csv('titanic_train.csv')

df.head(10)
```

```python
#Drop unesscary columns
df = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
df.head(20)
```

```python
df.isnull().any()
```

```python
#Fill in missing values
mean_value=df['Age'].mean()
df['Age'].fillna(value=mean_value, inplace=True)
df['Embarked'].fillna('S', inplace = True)
df.head(20)
```

```python
# Convert categorical variables into numerical variables
df['Sex'].replace(['male', 'female'], [0, 1], inplace=True) # 0 means male. 1 means female.
df['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2], inplace=True) # 0 means S. 1 means C. 2 means Q
df.head(20)
```

```python
# Define Feature Vector
feature_columns = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
```

```python
# Extract features
X = df.loc[:, feature_columns]
X.shape
```

```python
# Extract label
y = df.loc[:, "Survived"]
y.shape
```

```python
# Convert them to Numpy Arrays
X_arr = X.to_numpy()
y_arr = y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split

#Split Data
X_train, X_test, y_train, y_test = train_test_split(X_arr, y_arr, test_size=0.33, random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest Classifier object
clf = RandomForestClassifier(random_state=0)

# Train model
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)
```

```python
# Evaluate Random Forest Classifier model
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
f = open("output_titanic.txt", "w")
f.write("Classifcation Report for Random Forest\n")
f.write(report)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy is :")
print(accuracy_score(y_pred, y_test)*100) 
f.write("Accuracy is : {}\n\n".format(accuracy_score(y_pred, y_test)*100))
```
